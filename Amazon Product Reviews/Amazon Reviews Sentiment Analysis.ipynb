```
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```
import pandas as pd
data = pd.read_csv('amazon_reviews.tsv',sep="\t")   ## \t - tab seperated
data.head()
```

```
len(data)
```

```
data['label'].unique()
```

```
data['label'].value_counts()
```

```
polarity_score = []
for i in data['review']:
  polarity_score.append(sia.polarity_scores(i))
data['Polarity Score'] = polarity_score
```

```
data.head(20)
```

```
sentiment = []
for i in data['Polarity Score']:
  c = i['compound']
  if c>=0:
    sentiment.append("pos")
  else:
    sentiment.append("neg")
```

```
data['Sentiment']  = sentiment
data.head(20)
```

```
## Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(data['label'],data['Sentiment'])
cm
```

```
## Accuracy
from sklearn.metrics import accuracy_score
a = accuracy_score(data['label'],data['Sentiment'])
a
```
