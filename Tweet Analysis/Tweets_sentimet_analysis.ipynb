```
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```
import pandas as pd
data = pd.read_csv("twitter_data.csv",encoding='latin-1') 
```

```
polarity_score = []
for i in data["SentimentText"]:
  polarity_score.append(sia.polarity_scores(i))
data["Polarity Score"] =polarity_score
```

```
data.head(20)
```

```
compound = []
for i in data["Polarity Score"]:
  c = i["compound"]
  if c >= 0:
    compound.append("pos")
  else:
    compound.append("neg") 
```

```
data["Sentiment Analysis"] = compound
data.head(20)
```
